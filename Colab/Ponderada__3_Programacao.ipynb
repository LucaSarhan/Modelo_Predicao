```python
drop = ['tracao_animal',
        'trator_maquinas',
        'n_da_ocorrencia',
        'sentido',
        'tipo_de_acidente',
        'automovel',
        'transporte_de_cargas_especiais',
        'ilesos',
        'levemente_feridos',
        'moderadamente_feridos',
        'gravemente_feridos',
        'mortos',
        'tipo_de_ocorrencia',
        'horario']
df = df.drop(columns = drop, axis=1)
df
```

```python
df['lugar_acidente'].value_counts()
```

```python
dummies_place = pd.get_dummies(df['lugar_acidente'])
df = pd.concat([df, dummies_place], axis=1)
df = df.drop(columns =['lugar_acidente'])
df
```

```python
df = df.fillna(0.0)
df
```

```python
province = df['trecho'].str.split('/').str[1]
province = province.str.split(' ').str[0]
province = province.str.split('-').str[0]
df['trecho']= province
df = df.rename(columns={'trecho': 'estado'})
```

```python
dummies_province = pd.get_dummies(df['estado'])
df = pd.concat([df, dummies_province], axis=1)
df = df.drop(columns =['estado'])
df
```

```python
# A ideia era subitrair a data di acidente pela ultima ocorrência, entretando como acontece acidente todo dia, a coluna teria na sua totalidade 0.
# first_ocurrence = df['data'].iloc[0]
# first_ocurrence
# df_gab = pd.DataFrame()
# df_gab['last_ocurrence'] = pd.to_datetime(df['DataAcidente'])

# for data in df['data']
#     df_gab['last_ocurrence'] = df['data'] - first_ocurrence
#     first_ocurrence = df['data']
```

```python
# Numero de acidentes
df['accidents'] = df.index
# Crie uma coluna 'acidents_per_day' que conta o número de acidentes por dia
df['data'] = pd.to_datetime(df['data'])
df['acidents_per_day'] = df.groupby(df['data'].dt.date)['accidents'].transform('count')

df = df.drop(columns='data')
```

```python
# Normalização apenas da coluna 'km', number accidents e acidentes per day
min_km = df['km'].min()
max_km = df['km'].max()
min_accidents = df['accidents'].min()
max_accidents = df['accidents'].max()
min_acidents_per_day = df['acidents_per_day'].min()
max_acidents_per_day = df['acidents_per_day'].max()

# Normalização
df['km'] = (df['km'] - min_km) / (max_km - min_km)
df['accidents'] = (df['accidents'] - min_accidents) / (max_accidents - min_accidents)
df['acidents_per_day'] = (df['acidents_per_day'] - min_acidents_per_day) / (max_acidents_per_day - min_acidents_per_day)
```

```python
# Calcula a matriz de correlação para o DataFrame df e aplica um estilo de gradiente de cores
matriz_corr = df.corr().style.background_gradient(cmap='coolwarm')
matriz_corr
```

```python
# Renomear colunas para padrão internacional, sem pontuações e acentos
df.columns = [unidecode(col) for col in df.columns]

# Dataset é enorme, diminuir por causo que estavam timeout tanto no colab quanto no pycaret
df = df.head(30000)
```
